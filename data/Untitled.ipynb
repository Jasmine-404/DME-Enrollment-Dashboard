```python
# Step 1: 导入必要的库
import geopandas as gpd

# Step 2: 读取你的 GeoJSON 和 SHP 文件
points = gpd.read_file("glm_draft_output.geojson")      # 替换为你的点文件路径
wards = gpd.read_file("Wards_from_2022\Wards_from_2022.shp")             # 替换为你的面文件路径

# Step 3: 确保两个图层坐标系一致（建议统一为 EPSG:4326）
points = points.to_crs(epsg=4326)
wards = wards.to_crs(epsg=4326)
```

```python
# Step 4: 空间连接（spatial join）：找出每个点在哪个 ward 面内
joined = gpd.sjoin(points, wards[['NAME', 'geometry']], how='left', op='within')

# Step 5: 重命名 name 字段为 ward
joined = joined.rename(columns={'NAME': 'ward'})

# Step 6: 去掉不需要的列（可选）
joined = joined.drop(columns=['index_right'])

# Step 7: 输出新的 GeoJSON 文件
joined.to_file("glm_draft_output_new.geojson", driver='GeoJSON')

# 也可以直接显示部分结果看看效果
joined.head()
```
